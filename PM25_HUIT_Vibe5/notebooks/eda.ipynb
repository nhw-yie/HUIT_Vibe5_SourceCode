```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu sau khi đã merge (dataset_features.csv hoặc dataset_features_2.csv)
df = pd.read_csv("../data_processed/dataset_features.csv", parse_dates=["ts_utc"])

# 1. Histogram PM2.5
plt.figure(figsize=(6,4))
plt.hist(df["PM25"], bins=30, color="skyblue", edgecolor="black")
plt.xlabel("PM2.5 (µg/m³)")
plt.ylabel("Tần suất")
plt.title("Phân phối nồng độ PM2.5")
plt.show()

# 2. Heatmap tương quan
plt.figure(figsize=(10,8))
corr = df[["PM25","t2m_C","d2m_C","RH","ws","aod_mean"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Tương quan giữa PM2.5 và các biến khí tượng/AOD")
plt.show()

# 3. Line chart theo thời gian
plt.figure(figsize=(12,5))
plt.plot(df["ts_utc"], df["PM25"], label="PM2.5", color="red", linewidth=1)
plt.xlabel("Thời gian")
plt.ylabel("PM2.5 (µg/m³)")
plt.title("Biến động PM2.5 theo thời gian")
plt.legend()
plt.show()
```
