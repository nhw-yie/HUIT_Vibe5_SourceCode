```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# c d liu sau khi  merge (dataset_features.csv hoc dataset_features_2.csv)
df = pd.read_csv("../data_processed/dataset_features.csv", parse_dates=["ts_utc"])

# 1. Histogram PM2.5
plt.figure(figsize=(6,4))
plt.hist(df["PM25"], bins=30, color="skyblue", edgecolor="black")
plt.xlabel("PM2.5 (g/m)")
plt.ylabel("Tn sut")
plt.title("Phn phi nng  PM2.5")
plt.show()

# 2. Heatmap tng quan
plt.figure(figsize=(10,8))
corr = df[["PM25","t2m_C","d2m_C","RH","ws","aod_mean"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Tng quan gia PM2.5 v cc bin kh tng/AOD")
plt.show()

# 3. Line chart theo thi gian
plt.figure(figsize=(12,5))
plt.plot(df["ts_utc"], df["PM25"], label="PM2.5", color="red", linewidth=1)
plt.xlabel("Thi gian")
plt.ylabel("PM2.5 (g/m)")
plt.title("Bin ng PM2.5 theo thi gian")
plt.legend()
plt.show()
```
